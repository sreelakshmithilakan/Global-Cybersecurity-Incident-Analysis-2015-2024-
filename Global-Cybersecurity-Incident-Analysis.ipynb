### _Creating derived fields:_

```python
import numpy as np

# Column names
YEAR_COL = "Year"
LOSS_COL = "Financial Loss (in Million $)"
USERS_COL = "Number of Affected Users"
RES_HR_COL = "Incident Resolution Time (in Hours)"

# 1) Era: Pre vs Post-Pandemic (cut at 2020)

df['Era'] = None
df.loc[df['Year'] <= 2019, 'Era'] = 'Pre-Pandemic'
df.loc[(df['Year'] > 2019) & (df['Year'] <= 2021), 'Era'] = 'Pandemic (2020–2021)'
df.loc[(df['Year'] > 2021) & (df['Year'] <= 2023), 'Era'] = 'Recovery (2022–2023)'
df.loc[df['Year'] > 2023, 'Era'] = 'Recent (2024–2025)'

df['Era'] = df['Era'].astype('category')
print(df['Era'].value_counts(normalize=True) * 100)
print('_______________________________________________________________________________________________')

# 2) Loss per User ($) — avoid divide by zero

df["Loss per User ($)"] = ((df[LOSS_COL] * 1_000_000) / df[USERS_COL]).round(2) 
print(df["Loss per User ($)"].value_counts(normalize=True) * 100)
print('_______________________________________________________________________________________________')

# 3) Resolution Bucket — categorize resolution time

df['Resolution Bucket'] = df['Incident Resolution Time (in Hours)'].apply(
    lambda x: '0–24h' if x <= 24 else
              '24–48h' if x <= 48 else
              '48–72h' if x <= 72 else
              '>72h'
).astype('category')
print(df['Resolution Bucket'].value_counts(normalize=True) * 100)
print('_______________________________________________________________________________________________')


print(df[["Era", "Loss per User ($)", "Resolution Bucket"]].head())
```

### _Finding outliers using Interquartile Range Method:_

```python
print('Finding outliers using Interquartile Range Method:')
print('_______________________________________________________________________________________________')

num_cols = ['Financial Loss (in Million $)','Number of Affected Users','Incident Resolution Time (in Hours)','Loss per User ($)']

outlier_summary = {}

for col in num_cols:

    df[col] = pd.to_numeric(df[col], errors='coerce')


    col_min = df[col].min()
    col_max = df[col].max()
    col_range = col_max - col_min


    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    count = len(outliers)

    outlier_summary[col] = {
        "min": col_min,
        "max": col_max,
        "range": col_range,
        "IQR": IQR,
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "outliers_count": count
    }


for col, stats in outlier_summary.items():
    print(f"\n Column: {col}")
    print(f"   Min = {stats['min']}")
    print(f"   Max = {stats['max']}")
    print(f"   Range = {stats['range']}")
    print(f"   IQR = {stats['IQR']:.2f}")
    print(f"   Lower Bound = {stats['lower_bound']:.2f}")
    print(f"   Upper Bound = {stats['upper_bound']:.2f}")
    print(f"   Outliers Count = {stats['outliers_count']}")

print('''\nFor 'Loss per User ($)' 338 outliers are detected.
      
Here IQR is very small compared to your max. A majority of incidents probably involve low per-user losses (0–300 USD range). 

But a small number of incidents had very large losses per user (hundreds to thousands).

This stretches your range but doesn’t affect Q1 and Q3 much — they stay low, so the upper bound remains small → lots of outliers.

Conclusion: Thus the distribution is right-skewed (long tail on the higher side). A small percent of incidents are very costly per user.''')
```

### _Variability, Skewness & Standard Deviation of data:_

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Variability, Skewness & Standard Deviation in Data\n")
print("=====================================================================================")

for col in num_cols:
    print(f"\n Column: {col}")
    print("-" * (len(col) + 12))
    
    #Variability 
    col_min = df[col].min()
    col_max = df[col].max()
    minmax_diff = col_max - col_min
    print(f"Range (Max - Min): {minmax_diff:.2f}")
    
    if minmax_diff > df[col].median() * 2:
        print("Conclusion: High variability — wide spread in data.")
    else:
        print("Conclusion: Low variability — values are close together.")
    
    #Skewness 
    mean_val = df[col].mean()
    median_val = df[col].median()
    skewness = mean_val - median_val
    skew_type = df[col].skew()
    print(f"\nMean - Median = {skewness:.2f}")
    print(f"Skewness value = {skew_type:.2f}")
    
    if skew_type > 0.5:
        print("Conclusion: Right (positively) skewed distribution.")
    elif skew_type < -0.5:
        print("Conclusion: Left (negatively) skewed distribution.")
    else:
        print("Conclusion: Roughly symmetric distribution.")
    
    #Standard Deviation
    std_dev = df[col].std()
    print(f"\nStandard Deviation = {std_dev:.2f}")
    
    cv = (std_dev / mean_val) * 100 if mean_val != 0 else np.nan
    if cv > 100:
        print("Conclusion: Very high variability relative to mean.")
    elif cv > 50:
        print("Conclusion: Moderately variable data.")
    else:
        print("Conclusion: Data is fairly consistent.")
    
    print("\n" + "-"*80)
```

__Evolution of User Impact Across Industries | Year, Target Industry & Number of Affected Users | Heatmap__
 

```python

impact_table = df.groupby(['Year', 'Target Industry'])['Number of Affected Users'].sum().reset_index()

impact_pivot = impact_table.pivot(index='Target Industry', columns='Year', values='Number of Affected Users')

plt.figure(figsize=(14,8))
sns.heatmap(impact_pivot,cmap='YlOrRd',linewidths=0.5,linecolor='white',annot=True,fmt='.0f')

plt.title('Evolution of User Impact Across Industries', fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Target Industry')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

___How to read this heatmap:___

Each row → a specific industry (e.g., Banking, Healthcare, Energy).

Each column → a year.

Color intensity → the total number of affected users.

🔸 Lighter color → fewer users impacted.

🔴 Darker color → more users impacted.

So, dark red cells point to years and industries with major user impact.


Interpretation: 

- The heatmap reveals that IT and Education sectors consistently suffered the highest user impact, peaking in 2017 and 2018 respectively.
- Banking and Healthcare maintained steady vulnerability, reflecting their critical role in data-driven operations.
- Noticeable spikes around 2020–2021 across industries align with the global increase in cyber activity during the pandemic, highlighting the strain on digital infrastructures.
- Overall, the data underscores the need for sector-specific, adaptive cybersecurity strategies to manage evolving user risks.

- The comparison between pre-pandemic and post-pandemic periods shows that most attack types such as DDoS, Malware, Phishing, and Man-in-the-Middle maintained similar incident frequencies across both eras.

- This pattern highlights how cybercriminals adapted to the changing digital environment, leveraging increased online dependency and exposed endpoints after 2020.

```python

agg = (df.groupby(['Country', 'Target Industry', 'Attack Source']).size().reset_index(name='incidents'))

n_countries = 6
top_countries = (agg.groupby('Country')['incidents'].sum().nlargest(n_countries).index)
agg_sub = agg[agg['Country'].isin(top_countries)].copy()

agg_sub['Attack Source'] = agg_sub['Attack Source'].astype(str)
agg_sub['Target Industry'] = agg_sub['Target Industry'].astype(str)

g = sns.relplot(data=agg_sub,x='Attack Source',y='Target Industry',size='incidents',sizes=(50, 2000),hue='incidents',palette='viridis',
    col='Country',col_wrap=3,alpha=0.7,edgecolor='k',linewidth=0.3,kind='scatter',height=4,aspect=1.2,)

for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=45, labelsize=9)
    ax.set_xlabel('')
    ax.set_ylabel('')

g.set_axis_labels("Attack Source", "Target Industry")
g.fig.suptitle('Global Overview of Attack Sources Across Industries (Bubble Size = Incidents)', y=1.03, fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## **9. Temporal and Trend Analysis**

> ***Purpose:*** *Understand how threats and defenses have evolved from 2015 to 2024.*

__Evolution of Attack Landscape | Year & Attack Type | 100% Stacked Area Chart__

```python
attack_trend = df.groupby(["Year", "Attack Type"]).size().unstack(fill_value=0)

attack_percent = attack_trend.div(attack_trend.sum(axis=1), axis=0) * 100

attack_percent.plot.area(figsize=(10, 6), alpha=0.8)

plt.title("Evolution of Attack Landscape: Year & Attack Type", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Percentage Share (%)")
plt.legend(title="Attack Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- Between 2015 and 2024, the global attack landscape remained diverse and dynamic, with no single technique dominating consistently.
- Phishing and Ransomware emerged as enduring threats, while Malware and DDoS maintained stable shares.
- This stability reflects both attacker adaptability and improved multi-layered defense strategies.

__Defense Mechanism Effectiveness Over Time | Year & Avg Financial Loss (by Defense Mechanism) | Multi-series Line Chart__

```python
loss_trend = (df.groupby(["Year", "Defense Mechanism Used"])["Financial Loss (in Million $)"].mean().unstack())

loss_trend.plot(figsize=(10, 6), marker='o')

plt.title("Average Financial Loss by Defense Mechanism Over Time", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Avg Financial Loss (Million $)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title="Defense Mechanism", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

- Between 2015 and 2024, average financial losses per defense mechanism remained relatively stable, indicating that cyber defenses have matured but not dramatically improved efficiency.
- Encryption and modern AI-based detection show gradual improvement, while traditional defenses (firewall, antivirus) show stagnation.
- Post-2020 volatility aligns with pandemic-related shifts in network security patterns (VPN reliance, remote access).

__Global Cyber Impact Over Time | Financial Loss & Users | Dual-axis Line__

```python
plt.figure(figsize=(10,5))
yearly = df.groupby('Year')[['Financial Loss (in Million $)', 'Number of Affected Users']].sum().reset_index()

ax1 = plt.gca()
ax2 = ax1.twinx()

ax1.plot(yearly['Year'], yearly['Number of Affected Users'], 'o-', color='orange', label='Affected Users')
ax2.plot(yearly['Year'], yearly['Financial Loss (in Million $)'], 's-', color='blue', label='Financial Loss (M$)')

ax1.set_xlabel('Year')
ax1.set_ylabel('Affected Users')
ax2.set_ylabel('Financial Loss (Million $)')
ax1.set_title('Yearly Trend: Financial Loss & Affected Users (2015–2024)')

plt.axvline(2020, color='gray', ls='--', lw=1)
plt.tight_layout()
plt.show()
```

- Cyber incidents surged dramatically post-2020, peaking around 2021–2022, before stabilizing.
- This reflects a shift from pandemic-induced cyber chaos to a more controlled cybersecurity landscape, though risks remain significant.

## **11. Limitations & Future Scope**

### **Limitations**
- The dataset may not capture all global incidents, leading to regional or reporting bias.  
- Severity metrics (financial loss, user impact) depend on available data; some values might be estimated or incomplete.  
- Analysis assumes consistent reporting standards across countries and industries, which may not hold true in practice.  
- Temporal trends (especially around 2020–2021) could be affected by underreporting during the pandemic.  
- Defense effectiveness is evaluated from incident outcomes only, not from proactive prevention data, which limits insight into early threat mitigation.



### **Future Scope**
- Integrate real-time cybersecurity feeds or threat intelligence APIs to enhance trend accuracy.  
- Extend analysis to include economic and organizational attributes (e.g., company size, sector investment in security).  
- Apply machine learning models for predictive insights. e.g., forecasting financial loss or user impact from historical data.  
- Use network graph analysis to visualize global attack connections and propagation patterns.
